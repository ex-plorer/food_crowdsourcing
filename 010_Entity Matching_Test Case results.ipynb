```python
file_name = "food_entities__20180321.p"
df = f.retrive_file(file_name)
# index - assigned name ref
ref_assigned_index = df.to_dict()["assigned_name"]
# menu_name - index ref
ref_menu_index = df.to_dict()["menu_name"]
rev_menu_index = dict()
for k, vals in ref_menu_index.items():
    for val in vals:
        rev_menu_index.update({val[0]:k})
```

** positive cases **

```python
file_name = "002_entity matching_food_test_case_positive_20180321.csv"
import pandas as pd
tests = pd.read_csv(file_name)

# get index
tests["index_0"] = tests["food_0"].apply(lambda s: rev_menu_index[s])
tests["index_1"] = tests["food_1"].apply(lambda s: rev_menu_index[s])
# add food name
tests["index_0_name"] = tests["index_0"].apply(lambda s: ref_assigned_index[s])
tests["index_1_name"] = tests["index_1"].apply(lambda s: ref_assigned_index[s])
# compare index
tests["test_results"] = tests.apply(lambda l: l["index_0"] == l["index_1"], axis=1)

file_name = "002_entity_matching_test_case_"+f.today()+".csv"
tests.to_csv(file_name, index=False)
print "saved: %s"%file_name

from __future__ import division
round(tests["test_results"].sum() / tests["test_results"].shape[0],5)
```
